## Upsert New notices

```python
from sqlalchemy import create_engine, Column, String, Integer, Float, DateTime, Text, Table, MetaData
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy.orm import sessionmaker
import pandas as pd
import os
from datetime import datetime
from sqlalchemy import create_engine, Column, String, Integer, Float, DateTime, Text, Table, MetaData
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy.orm import sessionmaker
# Alternative implementation using pandas more directly
def pandas_direct_upsert(df_new, db_path='notices.db'):
    """
    A more straightforward upsert using pandas and SQLite directly.
    This is more efficient when all fields are treated as strings and
    we're doing a full record comparison.
    
    Args:
        df_new: DataFrame with new records
        db_path: Path to SQLite database
    
    Returns:
        Dictionary with statistics
    """
    stats = {'total': len(df_new), 'inserted': 0, 'updated': 0, 'errors': 0}
    
    try:
        # Connect to database
        engine = create_engine(f'sqlite:///{db_path}')
        
        # Convert all columns to string
        df_clean = df_new.copy()
        for col in df_clean.columns:
            df_clean[col] = df_clean[col].fillna('').astype(str)
        
        # Add timestamp columns
        now = datetime.utcnow().isoformat()
        df_clean['created_at'] = now
        df_clean['updated_at'] = now
        
        # Read existing data
        try:
            existing_df = pd.read_sql_table('notices', engine)
            for col in existing_df.columns:
                existing_df[col] = existing_df[col].fillna('').astype(str)
        except:
            # Table doesn't exist
            existing_df = pd.DataFrame(columns=df_clean.columns)
        
        # Get all columns for comparison (excluding id, created_at, updated_at)
        compare_columns = [col for col in df_clean.columns 
                           if col not in ('id', 'created_at', 'updated_at')]
        
        if not existing_df.empty:
            # Find records that already exist (exact match on all columns)
            # This creates a dataframe with indicators of matching rows
            merged = df_clean.merge(
                existing_df[compare_columns], 
                on=compare_columns,
                how='left',
                indicator=True
            )
            
            # Filter to get only new records
            new_records = merged[merged['_merge'] == 'left_only'].drop('_merge', axis=1)
            stats['inserted'] = len(new_records)
            stats['updated'] = len(df_clean) - len(new_records)
            
            # Only insert records that don't exist
            if not new_records.empty:
                new_records.to_sql('notices', engine, if_exists='append', index=False)
        else:
            # No existing records, insert all
            df_clean.to_sql('notices', engine, if_exists='append', index=False)
            stats['inserted'] = len(df_clean)
        
        print(f"Successfully processed {stats['total']} records:")
        print(f" - {stats['inserted']} inserted")
        print(f" - {stats['updated']} updated (no changes needed)")
        print(f" - {stats['errors']} errors")
            
    except Exception as e:
        print(f"Error during pandas upsert: {str(e)}")
        stats['errors'] = len(df_new)
        raise
        
    return stats

```

```python
import os
import boto3
import pandas as pd
from tqdm import tqdm
import requests



# 1) Get list of PDF files from local directory
pdf_files = [os.path.join(download_directory, f) for f in os.listdir(download_directory) 
             if f.lower().endswith('.pdf')]

# 2) Check which files already exist in S3
existing_files = []
paginator = aws_client.get_paginator('list_objects_v2')
pages = paginator.paginate(Bucket=bucket_name, Prefix=folder_prefix)

for page in paginator.paginate(Bucket=bucket_name, Prefix=folder_prefix):
    if 'Contents' in page:
        for obj in page['Contents']:
            existing_files.append(os.path.basename(obj['Key']))

# Filter out files that already exist in S3
files_to_upload = []
for local_path in pdf_files:
    filename = os.path.basename(local_path)
    if filename not in existing_files:
        files_to_upload.append(local_path)

print(f"Found {len(pdf_files)} PDF files locally")
print(f"Found {len(existing_files)} existing files in S3")
print(f"Will upload {len(files_to_upload)} new files")
```

```python
import os
import boto3
import pandas as pd
from tqdm import tqdm
import requests

# <<< assume df_to_upload is already defined >>>



# 3) Create folder (no progress needed for single operation)
# aws_client.put_object(Bucket=bucket_name, Key=older_prefix)


# 4) Upload with progress tracking
uploaded = []
success_count = 0
upload_failures = 0
access_failures = 0

with tqdm(total=len(files_to_upload), desc="Uploading new PDFs", unit='file') as pbar:
    for local_path in files_to_upload:
        filename = os.path.basename(local_path)
        print(f"Uploading {filename}...")
        s3_key = f"{folder_prefix}{filename}"
        file_status = {'file': filename, 's3_key': s3_key}

        try:
            # Upload to S3
            aws_client.upload_file(
                Filename=local_path,
                Bucket=bucket_name,
                Key=s3_key,
                ExtraArgs={'ACL': 'private', 'ContentType': 'application/pdf'}
            )
            
            # Test CloudFront access
            cf_url = f"https://{cloudfront_domain}/{s3_key}"
            file_status['cloudfront_url'] = cf_url
            
            try:
                response = requests.head(cf_url, timeout=10)
                if response.status_code == 200:
                    file_status['status'] = "✅ Success"
                    success_count += 1
                else:
                    file_status['status'] = f"❌ HTTP {response.status_code}"
                    access_failures += 1
            except Exception as e:
                file_status['status'] = f"❌ Access Error: {str(e)}"
                access_failures += 1

        except Exception as e:
            file_status['status'] = f"❌ Upload Error: {str(e)}"
            upload_failures += 1

        uploaded.append(file_status)
        pbar.set_postfix({
            'OK': success_count,
            'UP Fail': upload_failures,
            'CF Fail': access_failures
        })
        pbar.update(1)

# 5) Create results DataFrame
uploaded_df = pd.DataFrame(uploaded)

# 6) Print summary
print("\nUpload Summary:")
print(f"Total files processed: {len(files_to_upload)}")
print(f"Successfully uploaded & accessible: {success_count}")
print(f"Upload failures: {upload_failures}")
print(f"CloudFront access failures: {access_failures}")
```
